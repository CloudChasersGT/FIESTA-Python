### Gridcells Spatial Files 

```python
elev_range = [1350, 2650]
elev_step = 100
slope_range = [5, 65]
slope_step = 10
aspect_range = [22.5, 337.5]
aspect_step = 45

elevs = np.arange(elev_range[0], elev_range[1]+elev_step, elev_step)
aspects = np.arange(aspect_range[0], aspect_range[1]+aspect_step, aspect_step)
slopes = np.arange(slope_range[0], slope_range[1]+slope_step, slope_step)

land_class = np.zeros((len(elevs) * len(aspects) * len(slopes), 3))

s = 0
for i in range(len(elevs)): 
    land_class[s:(s+len(aspects)*len(slopes)), 0] = elevs[i]
    p = s
    for j in range(len(aspects)):
        land_class[p:(p+len(slopes)), 1] = aspects[j] 
        for k in range(len(slopes)): 
            land_class[p+k, 2] = slopes[k]
        p = p + len(slopes)
    s = s + len(aspects)*len(slopes)
np.savetxt('spatialdist.csv', land_class, delimiter=',')



#Scn1
# Frac_clr    = Gridcells[:,5]*2
# Frac_frst   = Gridcells[:,6]*0

#Scn2
# Frac_clr    = Gridcells[:,5]*1
# Frac_frst   = Gridcells[:,6]*1

# #Scn3 
# Frac_clr    = Gridcells[:,5]*1.5
# Frac_frst   = Gridcells[:,6]*.5

# #Scn4
# Frac_clr    = Gridcells[:,5]*.5
# Frac_frst   = Gridcells[:,6]*1.5
```
